# 🍽️ Cognifyz ML Internship — Task 4
## Location-Based Analysis of Restaurants
**Objective:** Perform a geographical analysis of the restaurants in the Zomato dataset.

**Steps:**
1. Explore latitude & longitude and visualize on a map
2. Group restaurants by city and analyze concentration
3. Calculate average ratings and price ranges by city
4. Identify interesting insights from location data


## 📦 Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
```

## 📂 Step 2: Load Dataset

```python
url = "https://raw.githubusercontent.com/vp763/ML_Project_cognifyz/refs/heads/main/Dataset%20.csv"
df = pd.read_csv(url)
print(f"Dataset shape: {df.shape}")
df.head()
```

## 🔍 Step 3: Data Exploration

```python
df.info()
```

## 🧹 Step 4: Data Cleaning
> **Missing Values:** Cuisines column has 9 missing values (0.09% of data). Dropping them since it is negligible.

```python
# Drop rows with missing Cuisines (only 9 rows — negligible)
df.dropna(subset=["Cuisines"], inplace=True)
print(f"Dataset shape after cleaning: {df.shape}")
```

### Encode Categorical Columns
> Converting Yes/No columns to 1/0 since ML models understand numbers, not text.

```python
# Binary encoding: Yes -> 1, No -> 0
df["Has Table booking"]   = df["Has Table booking"].map({"Yes": 1, "No": 0})
df["Has Online delivery"] = df["Has Online delivery"].map({"Yes": 1, "No": 0})
df["Is delivering now"]   = df["Is delivering now"].map({"Yes": 1, "No": 0})

# Verify encoding
df[["Has Table booking", "Has Online delivery", "Is delivering now"]].value_counts()
```

## 📊 Step 5: City-wise Restaurant Analysis

### 5a. Top 10 Cities by Number of Restaurants

```python
top_cities = df["City"].value_counts().head(10)

plt.figure(figsize=(12, 6))
top_cities.plot(kind="bar", color="steelblue", edgecolor="black")
plt.title("Top 10 Cities by Number of Restaurants", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### 5b. Top 10 Cities by Average Rating

```python
avg_rating = df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_rating.plot(kind="bar", color="mediumseagreen", edgecolor="black")
plt.title("Top 10 Cities by Average Restaurant Rating", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### 5c. Top 10 Cities by Average Cost for Two

```python
avg_price = df.groupby("City")["Average Cost for two"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_price.plot(kind="bar", color="coral", edgecolor="black")
plt.title("Top 10 Cities by Average Cost for Two", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Average Cost (Local Currency)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## 🗺️ Step 6: Geographical Visualization
### 6a. Restaurant Location Map (Sample: 500 Restaurants)

```python
# Create base map centered on India
restaurant_map = folium.Map(location=[20.5, 78.9], zoom_start=5)

# Add markers for first 500 restaurants
for index, row in df.head(500).iterrows():
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=row["Restaurant Name"]
    ).add_to(restaurant_map)

restaurant_map
```

### 6b. Restaurant Density Heatmap (All Restaurants)

```python
# Heatmap shows concentration of restaurants across the world
heat_map = folium.Map(location=[20.5, 78.9], zoom_start=4)
heat_data = [[row["Latitude"], row["Longitude"]] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(heat_map)
heat_map
```

## 💡 Step 7: Key Insights

1. **New Delhi dominates** — With 5,473 restaurants, it has more listings than all other cities combined.
2. **Top rated cities** are not necessarily the most populated ones — smaller cities sometimes have higher average ratings.
3. **Cost varies significantly by city** — Some cities have dramatically higher average costs, reflecting economic differences.
4. **Geographical concentration** — The heatmap clearly shows restaurant density is highest in South Asia (India), with clusters visible in other regions like Middle East and Southeast Asia.
5. **NCR region dominates** — New Delhi, Gurgaon, Noida, Faridabad, and Ghaziabad together form a massive restaurant cluster in North India.
